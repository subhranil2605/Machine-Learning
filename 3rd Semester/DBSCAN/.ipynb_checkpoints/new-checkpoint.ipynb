```python
from __future__ import annotations

class Point:
    """
    Point class in a 2D plane
    """
    
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y
        
    def dist(self, other: Point):
        """
        Calculates the Euclidean distance between two points
        """
        return math.dist([self.x, self.y], [other.x, other.y])
    
    def __eq__(self, other: Point):
        """
        Checks if the two points are equal or not
        """
        return self.x == other.x and self.y == other.y
    
    def __repr__(self):
        """
        Represent the point object
        """
        return f"Point: ({self.x}, {self.y})"
    
    def __hash__(self):
        return hash(str(self))
```

```python
def is_core(p: Point, ds, eps: float, min_pts: int):
    """
    Checks if the point is the core point or not
    """
    # eps-neighborhood
    eps_nghd = [i for i in ds if p.dist(i) <= eps]
    
    # checks if the neighborhood has points
    # at least or greater than min_points then it is a core point
    return eps_nghd if len(eps_nghd) >= min_pts else None
```

```python
def DBSCAN(ds, eps: float, min_pts: int):
    visited = set()
    clusters = []
    
    for p in ds:
        if p not in visited:
            cluster = []
            eps_neghd = is_core(p, ds, eps, min_pts)
            if eps_neghd:                    
                cluster.extend(eps_neghd)

            clusters.append(cluster) if cluster else None
            visited.add(p)
            
    return clusters 
            
DBSCAN(df, 0.12, 3)
```
