## Point class

```python
from __future__ import annotations

class Point:
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y
    
    def dist(self, other: Point) -> float:
        """
        Calculates Euclidean distance between two points
        """
        return np.sqrt(np.power(self.x - other.x, 2) + np.power(self.y - other.y, 2))
    
    def __eq__(self, other: Point):
        """
        Checks if the two point objects are equal or not
        """
        return self.x == other.x and self.y == other.y
```

# Calculates all the core points from a dataset

```python
def core_points(ds: ndarray, eps: float, min_points: int):
    """
    Returns the core points
    """
    core_points = []
    d = []
    
    for i in tqdm(ds):
        eps_neighbourhood = []
        eps_neighbourhood_count = 0
        
        p = Point(i[0], i[1])
        for j in ds:
            q = Point(j[0], j[1])
            if p != q: # checks if the same point
                if p.dist(q) <= eps: # how many points are in the eps-radius circle
                    eps_neighbourhood.append(q)
                    eps_neighbourhood_count += 1
        
        d.append(eps_neighbourhood)
        
        if eps_neighbourhood_count >= min_points:
            core_points.append(p)
        
    return core_points, d
```

```python
eps = 0.4
min_points = 10

# eps = 22.5
# min_points = 4

cores = core_points(X_, eps, min_points)
```

```python
len(cores[0]), len(cores[1])
```

## Core Points Plot

```python
plt.scatter(X, Y)
plt.scatter([i.x for i in cores[0]], [i.y for i in cores[0]])
plt.title(f"Core Points\neps={eps} and min_points = {min_points}")
```

## Boundary Points

```python
def boundary_points(cores_, eps_nghs):
    b = set()
    for i in eps_nghs:
        if i:
            for j in i:
                if j not in cores_:
                    a = (j.x, j.y)
                    b.add(a)
    return b

        
b = boundary_points(cores[0], cores[1])
len(b)
```

```python
boundaries = [Point(i[0], i[1]) for i in b]
```

```python
plt.scatter(X, Y)
plt.scatter([i.x for i in boundaries], [i.y for i in boundaries], c='y')
plt.title(f"Boundary Points\neps={eps} and min_points = {min_points}")
```

## Noise

```python
outliers = []
for i in X_:
    p = Point(i[0], i[1])
    
    if p not in cores[0] and p not in boundaries:
        outliers.append(p)
```

```python
len(outliers)
```

```python
plt.scatter(X, Y)
plt.scatter([i.x for i in outliers], [i.y for i in outliers], c='k')
plt.title(f"Noise Points\neps={eps} and min_points = {min_points}")

plt.show()
```

## Combine: Core and Boundary: DBSCAN

```python
plt.scatter([i.x for i in outliers], [i.y for i in outliers], c='k', label="Noise Points")
plt.scatter([i.x for i in cores[0]], [i.y for i in cores[0]], c='r', label="Core Points")
plt.scatter([i.x for i in boundaries], [i.y for i in boundaries], c='y', label="Boundary Points")
plt.title(f"DBSCAN\neps={eps} and min_points = {min_points}")

plt.legend()
plt.show()
```
